```python

import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("photo.jpg")   # Load image
# cv2.imshow("My Image", img)
# cv2.waitKey(0)
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
# cv2.imshow("Gray", gray)
# cv2.waitKey(0)

# Useful
# cv2.COLOR_BGR2RGB
# cv2.COLOR_BGR2HSV
```

```python
# Resize an Image
rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
resized = cv2.resize(rgb, (500, 300))
# cv2.imshow("Resized", resized)
# cv2.waitKey(0)
plt.imshow(resized)
plt.axis('off')
plt.show()
```

```python
# scale by ratio
small = cv2.resize(rgb, None, fx=0.3, fy=0.3) 
plt.imshow(small)
plt.axis('off')
plt.show()
```

```python
crop = img[100:200, 200:400] # [y1:y2, x1:x2]
plt.imshow(crop)
plt.axis('off')
plt.show()
```

```python
line = cv2.line(img, (50, 50), (300, 50), (0, 0, 255), 3)
# cv2.imshow("My Image", img)
# cv2.waitKey(0)
plt.imshow(line)
plt.show()

line = cv2.line(rgb, (50, 50), (300, 50), (0, 0, 255), 3)
# cv.rectangle
circle = cv2.circle(rgb, (200, 200), 50, (255, 0, 0), -1)

plt.imshow(circle)
plt.show()


# If using matplotlib i.e plt.show(), evrything will change to RGB
# But when we use cv2.imshow(), original image will be as is as it was imported using cv2
# And we are drawing line using cv2, and showing it using plt so it uses --> RGB <-- format and gives blue color
# plt : RGB ****  cv2 : BGR
```

```python
text = cv2.putText(rgb, "Hello CV2", (50, 400),
            cv2.FONT_HERSHEY_SIMPLEX, 3, (255,255,255), 2)

plt.imshow(text)
plt.show()
```

```python
# Blurring the image

blur = cv2.GaussianBlur(rgb, (15, 15), 0)
plt.imshow(blur)
plt.show()

# (15, 15): The Kernel Size. It must be an odd, positive integer.
# Larger numbers (like 31, 31) create a much heavier blur.
# Smaller numbers (like 3, 3) provide a subtle cleanup.
# 0: The SigmaX (standard deviation in the X direction). When set to 0, OpenCV automatically calculates it based on your kernel size''' """
```

```python
# Edge Detection
edges = cv2.Canny(blur, 100, 200)
plt.imshow(edges)
plt.show()
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
_, thresh = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)
# cv2.imshow('thresh', thresh)
# cv2.waitKey(0)

plt.imshow(thresh)
plt.show()
```
