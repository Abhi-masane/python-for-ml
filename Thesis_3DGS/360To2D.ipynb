```python
# Class to convert Panoramic images to 2D images

class PanoTo2D:

    """ Converts 360 degree equirectangular images to multiple 2D perspective views """
    
    def __init__(self, op_width=800, op_height=600, fov=90):

        self.width = op_width
        self.height = op_height
        self.fov = fov

    def camera_directions(self, num_views=8):

        """ Generate Camera Orientations around sphere
        
        Input: Num of 2D images to extract
        
        Returns: List of (h_rot, v_rot) in degrees 
        h_rot = horizontal rotation 
        v_rot = vertical rotation """
                
        directions = []

        # Cameras at eye level for 360 degrees
        for i in range(num_views):
            h_rot = (360 / num_views) * i
            v_rot = 0 # at eye level
            directions.append((h_rot, v_rot))


        # Adding upward and downward views
        directions.append((0, 45)) # up 45 degrees
        directions.append((0, -45)) # down 45 degrees

        return directions





```
